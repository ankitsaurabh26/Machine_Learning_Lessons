#### 2D Line plot
- Bivariate Analysis
- Categorical -> Numerical & Numerical -> Numerical
- Use case : Time series data

```python
# Plotting some simple function
price = [48000,54000,49000,47000,45000,57000,59000]
year = [2012,2015,2016,2017,2018,2019,2020]

plt.plot(year,price) # plt.plot(x_axis,y_axis) 
```

```python
# from a pandas dataframe
batsman = pd.read_csv('sharma-kohli.csv')
batsman.head()
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
```

```python
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.title('Rohit Sharma vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='red')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#2DEEEA')
plt.title('Rohit Sharma vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='dashed')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#2DEEEA',linestyle='dotted')
plt.title('Rohit Sharma vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# linewidth
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='dashdot')
plt.plot(batsman['index'],batsman['RG Sharma'],color='#2DEEEA',linestyle='dotted',linewidth=4)
plt.title('Rohit Sharma vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# marker, markersize
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='dashdot',marker='D') # D -> Diamond
plt.plot(batsman['index'],batsman['RG Sharma'],color='#2DEEEA',linestyle='dotted',marker='o',markersize=6)
plt.title('Rohit Sharma vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
# label -> legend -> loc
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='dashdot',marker='D', label='Virat') # D -> Diamond
plt.plot(batsman['index'],batsman['RG Sharma'],color='#2DEEEA',linestyle='dotted',marker='o',markersize=6,label='Rohit')
plt.title('Rohit Sharma vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend(loc='lower right') # loc='best' -> default
```

```python
# limiting axes : ylim, xlim
price = [48000,54000,49000,47000,45000,57000,59000,4500000]
year = [2012,2015,2016,2017,2018,2019,2020,2021]
plt.plot(year,price,color='#77C108')
plt.xlabel('year')
plt.ylabel('price')
plt.ylim(0,100000)
```

```python
# limiting axes : ylim, xlim
price = [48000,54000,49000,47000,45000,57000,59000,4500000]
year = [2012,2015,2016,2017,2018,2019,2020,2021]
plt.plot(year,price,color='#77C108')
plt.xlabel('year')
plt.ylabel('price')
plt.ylim(0,100000)
plt.xlim(2017,2019)
```

```python
# label -> legend -> loc
plt.plot(batsman['index'],batsman['V Kohli'],color='red',linestyle='dashdot',marker='D', label='Virat') # D -> Diamond
plt.plot(batsman['index'],batsman['RG Sharma'],color='#2DEEEA',linestyle='dotted',marker='o',markersize=6,label='Rohit')
plt.title('Rohit Sharma vs Virat Kohli Career Comparison')
plt.xlabel('Season')
plt.ylabel('Runs Scored')
plt.legend(loc='lower right') # loc='best' -> default
plt.grid()
plt.show()  # to show the graph in text editor or websites
```

#### Bar Chart
- Bivariate Analysis
- Numerical vs Categorical
- Use case : Aggregate analysis of groups

```python
children = [70,47,99,32,69]
colors = ['Red','Blue','Green','Yellow','Pink']
plt.bar(colors,children,color='Orange')
```

```python
# Horizontal Bar Chart : No of categories too much
plt.barh(colors,children,color='#7141FA')
```

```python
# color and label
btr = pd.read_csv('batsman_season_record.csv')
btr
```

```python
plt.bar(btr['batsman'],btr['2015'])
plt.bar(btr['batsman'],btr['2016']) # graph overlapped
```

```python
np.arange(btr.shape[0])
```

```python
# Jugaad method to plot bar side-by-side
plt.bar(np.arange(btr.shape[0]) - 0.2,btr['2015'], width = 0.2,color='Green')
plt.bar(np.arange(btr.shape[0]),btr['2016'], width = 0.2,color='Orange')
plt.bar(np.arange(btr.shape[0]) + 0.2,btr['2017'], width = 0.2,color='Pink')
plt.xticks(np.arange(btr.shape[0]), btr['batsman']) # plt.xticks(old_name, replaced_name)
plt.show()
```

#### Pie Chart
- Univariate/Bivariate Analysis
- Categorical vs Numerical
- Use Case : To find contribution on a standard scale

```python
n_data = [100,24,89,67,59]
subjects = ['English','Hindi','Urdu','Bengali','Kannada']
plt.pie(n_data, labels = subjects)
plt.show()
```

```python
universe_boss = pd.read_csv('gayle-175.csv')
universe_boss
```

```python
plt.pie(universe_boss['batsman_runs'],labels=universe_boss['batsman'])
plt.show()
```

```python
plt.pie(universe_boss['batsman_runs'],labels=universe_boss['batsman'],autopct='%0.1f%%')
plt.show()
```

```python
plt.pie(universe_boss['batsman_runs'],labels=universe_boss['batsman'],autopct='%0.1f%%',colors=['Red','Grey','Green','Pink','Yellow','Orange'])
plt.show()
```

```python
plt.pie(universe_boss['batsman_runs'],labels=universe_boss['batsman'],autopct='%0.1f%%',colors=['Red','Grey','Green','Pink','Yellow','Orange'],explode=[0.2,0,0,0,0,0.3])
plt.show()
```
