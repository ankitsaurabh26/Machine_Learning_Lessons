#### Annotations

```python
batters = pd.read_csv('batter.csv')
```

```python
batters.head(5)
```

```python
sample_df = batters.head(100).sample(25, random_state=5)
```

```python
sample_df
```

```python
plt.scatter(sample_df['avg'],sample_df['strike_rate'])
plt.show()
```

```python
x = [1,3,5,6]
y = [3,4,6,7]
plt.scatter(x,y)
plt.text(1,3,'Point-1')
plt.text(5,6,'Point-3',fontdict={'size':12,'color':'Red'})
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.scatter(sample_df['avg'],sample_df['strike_rate'])
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
plt.show()
```

```python
# Bubble size : s -> according to runs 
plt.figure(figsize=(15,7))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
plt.show()
```

#### Horizontal and Vertical Lines

```python
plt.figure(figsize=(15,7))
plt.scatter(sample_df['avg'],sample_df['strike_rate'],s=sample_df['runs'])
plt.axhline(130,color="red") 
plt.axvline(30,color="yellow") 
plt.axhline(140,color="Green") 
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
plt.show()
```

#### 3D Surface Plot

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx,yy=np.meshgrid(x,y)
```

```python
xx
```

```python
yy
```

```python
xx.shape
```

```python
yy.shape
```

```python
z = xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='winter')
fig.colorbar(p)
plt.show()
```

```python
z = np.sin(xx)-np.cos(yy)
fig = plt.figure(figsize=(12,8))
ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='summer')
fig.colorbar(p)
plt.show()
```

#### Contour Plots
- 3D Graph to 2D Graph

```python
z = np.sin(xx)-np.cos(yy)
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contour(xx,yy,z,cmap='viridis') # top view, color is third axis : elevation and depth
fig.colorbar(p)
plt.show()
```

```python
z = np.sin(xx)-np.cos(yy)
fig = plt.figure(figsize=(12,8))
ax = plt.subplot()
p = ax.contourf(xx,yy,z,cmap='viridis') # top view -> color is third axis : elevation and depth
fig.colorbar(p)
plt.show()
```

#### HeatMap

```python
delivery = pd.read_csv('IPL_Ball_by_Ball_2008_2022.csv')
delivery.sample(5)
```

```python
delivery ['ballnumber'].unique()
```

```python
tempdf = delivery[(delivery ['ballnumber'].isin([1,2,3,4,5,6])) & (delivery['batsman_run']==6)]
```

```python
grid = tempdf.pivot_table(index='overs',columns='ballnumber',values='batsman_run',aggfunc='count')
```

```python
grid
```

```python
# heatmap
plt.figure(figsize=(10,6))
plt.imshow(grid,cmap='winter')
plt.yticks(delivery['overs'].unique(),list(range(1,21)))
plt.xticks(np.arange(0,6),list(range(1,7)))
plt.colorbar()
plt.show()
```
